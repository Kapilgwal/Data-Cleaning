```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data_full_dirty.csv')
df
```

```python
df.info()
```

```python
df.drop(columns = ['Unnamed: 0'], axis=0,inplace=True)
```

```python
df
```

```python
def company_name(text):
    arr = text.split(' ')
    return arr[0]
```

```python
df['company'] = df['name'].apply(company_name)
```

```python
df
```

```python
def fetch_year(text):
    text = str(text)
    arr = text.split(',')
    return str(arr[0])
```

```python
fetch_year('2019,April')
```

```python
df['year_rel'] = df['year'].apply(fetch_year)
```

```python
df
```

```python
def fetch_month(text):
    text = str(text)
    arr = text.split(',')
    
    if(len(arr) > 1):
        return str(arr[1])
    
    else:
        return "nan"
```

```python
df['month'] = df['year'].apply(fetch_month)
```
